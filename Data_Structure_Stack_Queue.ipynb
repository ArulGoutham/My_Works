# Situation:

#  There are two persons A and B. Both are maintaining two book shelves. A takes the first book (the book he kept for the first time in the shelf) from the shelf and reads it. B prefers to take the last book and reads. After reading a book, both A and B keep the book in a separate shelf.

# Approach 1: Implementation of Stack and Queue

```python
class Shelf:
    def __init__(self): 
        self.A = Queue() # Shelf of A
        self.B = Stack() # Shelf of B
        self.C = Queue() # COMMON SHELF
        self.s_sizeA = 0 
        self.s_sizeB = 0 
    
    def Add_NewBook_To_Shelf(self, data1,data2): 
        self.s_sizeA += 1
        self.s_sizeB += 1 
        self.A.enqueue(data1) 
        self.B.Push(data2)
        
    
    def Take_Book_from_Shelf(self): 
        if self.A.is_empty() and self.B.isEmpty(self.B):
            print("Shelf A and Shelf B are EMPTY")
        elif self.A.is_empty():
            print("Shelf A is EMPTY")
            self.s_sizeB -= 1
            self.C.enqueue(self.B.Pop()) # B Takes and reads a book and keeps it in shelf C
        elif self.B.isEmpty(self.B):
            print("Shelf B is EMPTY")
            self.s_sizeA -= 1
            self.C.enqueue(self.A.dequeue()) # A Takes and reads a book and keeps it in shelf C
        else:
            self.s_sizeB -= 1 
            self.s_sizeA -= 1
            self.C.enqueue(self.A.dequeue()) # A Takes and reads a book and keeps it in shelf C
            self.C.enqueue(self.B.Pop()) # B Takes and reads a book and keeps it in shelf C
    
    def Display_Books_in_Shelf(self): 
        print("SHELF A")
        self.A.display_queue()
        print("SHELF B")
        self.B.Display() 
        print("COMMON SHELF ")
        self.C.display_queue()
    
```

```python
class Stack:
    def __init__(self): 
        self.stk = []  # to store the elements in a queue
        self.item = 0 
    
    def isEmpty(self,stk): # checks whether the stack is empty or not
        if stk==[]:
            return True
        else:
            return False

    def Push(self,item): # Allow additions to the stack
        self.stk.append(item)
        top=len(self.stk)-1

    def Pop(self):
        if self.isEmpty(self.stk): # verifies whether the stack is empty or not
            return
        else: # Allow deletions from the stack
            self.item=self.stk.pop()
            if len(self.stk)==0:
                top=None
            else:
                top=len(self.stk)-1
            return self.item
    def top(self):
        if self.isEmpty(self.stk):
            print("Stack is empty")        
        else:
            top=len(self.stk)-1
            return self.stk[top]

    def Display(self):
        if self.isEmpty(self.stk):
            print("NO BOOKS IN SHELF ")
        
        else:
            top=len(self.stk)-1
          
            for i in range(top,-1,-1):
                print (str(self.stk[i]))
```

```python
class Queue:
    def __init__(self): # a constructor
        self.items = []  # LIST TO STORE BOOKS
        self.q_size = 0 # to compute the size of the queue
        
    def is_empty(self): # checking whether a queue is empty or not
        return self.items == []
    
    def enqueue(self, data): # to insert a new data to the queue
        self.q_size += 1 # increasing the size of a queue when a new data is inserted
        self.items.append(data) # inserting new data to the queue list 
        
    def dequeue(self): # to remove an element from the queue
        if(self.q_size == 0):
            print()
        else:
            q1=self.items.pop(0)
            self.q_size -= 1 # reducing the size of the queue when an element is removed from the queue  
            return q1 # delete the items from the front. pop() uses the index 0 to delete the element from the front
    
    def get_size(self): # to get the size of the queue 
        return self.q_size
    
    def display_queue(self): # to print the elements in the queue
        if(self.q_size == 0):
            print("NO BOOKS IN SHELF ")
        else:
            print(self.items[0:self.q_size]) # return all the elements in the queue
```

```python
if __name__ == '__main__':
    
    NewShelf = Shelf() 
      
    while True:
        
        Choice=int(input("Enter your option: "
                         "1. To Add books to the shelves"
                         "2. To Read books "
                         "3. To exit ") )
        
        
        if Choice == 1 :
            BookA = input("Enter the BookA: ")
            BookB = input("Enter the BookB: ")
            NewShelf.Add_NewBook_To_Shelf(BookA,BookB)      
            NewShelf.Display_Books_in_Shelf()
            print("----------------------------\n")
            
        elif Choice == 2:
            NewShelf.Take_Book_from_Shelf()
            NewShelf.Display_Books_in_Shelf()
            print("----------------------------\n")
            
        elif Choice == 3:
            break
```

# Approach 2: Implementing Stack and Queue using List

```python
class Arrange_Shelf:
    
    def __init__(self):
        self.A = []  # Shelf A - maintained by person 'A'
        self.B = []  # Shelf B - maintained by person 'B'
        self.C = []  # Shelf C - maintained by both 'A' and 'B' after reading their books
        self.size = 0
 
    def add_book(self, data): # Adding the books to the shelf A & B
        self.size += 1
        self.A.append(data)
        self.B.append(data) 
    
    def add_book_to_A(self, data): # Adding to the shelf --- If person'A' has more books than person'B'
        self.size += 1
        self.A.append(data)
    
    def add_book_to_B(self, data): # Adding to the shelf --- If person'B' has more books than person'A'
        self.size += 1
        self.B.append(data)
    
    def take_book_for_reading(self):
        self.size -= 1
        if len(self.A) == 0 and len(self.B) == 0: 
            print("Sheves A & B are Empty!!!") 
        elif len(self.A) == 0: 
            self.C.append(self.B.pop(-1)) # B person takes the book from shelf B for reading and after reading, he keeps that book in shelf C.
            print("Shelf A is Empty!!!")
        elif len(self.B) == 0:
            self.C.append(self.A.pop(0)) # A person takes the book from shelf A for reading and after reading, he keeps that book in shelf C.
            print("Shelf B is Empty!!!")
        else:
            self.C.append(self.A.pop(0)) # A person takes the book from shelf A for reading and after reading, he keeps that book in shelf C.
            self.C.append(self.B.pop(-1)) # B person takes the book from shelf B for reading and after reading, he keeps that book in shelf C.
    
    def books_in_the_shelf_A(self):
        print(self.A)
    
    def books_in_the_shelf_B(self):
        print(self.B)
    
    def books_in_the_shelf_C(self):
        print(self.C)
        
```

```python
if __name__ == '__main__':
    
    Arrange = Arrange_Shelf()
    
    while True:
        option = input("Enter your option: 1. Adding book to sheves 2. A wants to add more books "
                       "3. B wants to add more books  4. Taking the book for reading "
                       "5. Going to buy new books(Quit) : ")
        option = int(option.lower())
        
        if option == 1:     
            val = input("Enter the value: ")
            Arrange.add_book(val)
            print("Shelf maintained by A ")
            Arrange.books_in_the_shelf_A()
            print("Shelf maintained by B ")
            Arrange.books_in_the_shelf_B()
            print("Seperate shelf maintained by A after reading ")
            Arrange.books_in_the_shelf_C()
            print("----------------------------\n")
        
        elif option == 2:     
            val = input("Enter the value: ")
            Arrange.add_book_to_A(val)
            print("Shelf maintained by A ")
            Arrange.books_in_the_shelf_A()
            print("Shelf maintained by B ")
            Arrange.books_in_the_shelf_B()
            print("Seperate shelf maintained by A after reading ")
            Arrange.books_in_the_shelf_C()
            print("----------------------------\n")
        
        elif option == 3:     
            val = input("Enter the value: ")
            Arrange.add_book_to_B(val)
            print("Shelf maintained by A ")
            Arrange.books_in_the_shelf_A()
            print("Shelf maintained by B ")
            Arrange.books_in_the_shelf_B()
            print("Seperate shelf maintained by A after reading ")
            Arrange.books_in_the_shelf_C()
            print("----------------------------\n")
        
        elif option == 4:
            Arrange.take_book_for_reading()
            print("Shelf maintained by A ")
            Arrange.books_in_the_shelf_A()
            print("Shelf maintained by B ")
            Arrange.books_in_the_shelf_B()
            print("Seperate shelf maintained by A after reading ")
            Arrange.books_in_the_shelf_C()
            print("----------------------------\n")
            
        elif option == 5:
            break
```
